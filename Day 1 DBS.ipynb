```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline
```

```python
df = pd.read_csv(r"C:\Users\MOJITO\Downloads\DBS_SingDollar.csv")
```

```python
X = df.loc[:,["SGD"]]
Y = df.loc[:,["DBS"]]
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
pred = model.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = mean_squared_error(Y,pred) ** 0.5
```

```python
rmse
```

```python
import matplotlib.pyplot as plt
plt.plot(X, Y)
plt.xlabel('parameter')
plt.ylabel('rmse')
plt.title('rmse')
plt.show()
```
